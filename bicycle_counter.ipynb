#### Data Analysis and Visualization
Fremont Bridge Bicycle Counter

```python
# set for easy download for later use
URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
from urllib.request import urlretrieve
import pandas as pd
```

```python
# use the set URL to retrieve the data and save as Fremont.csv
urlretrieve(URL, 'data/Fremont.csv')
```

```python
# read the csv file and investigate the data
data = pd.read_csv('data/Fremont.csv', index_col='Date', parse_dates=True)
```

```python
data.head()
```

```python
# let us visualize this data
# import matplotlib.pyplot as plt
%matplotlib inline
data.plot()
```

```python
# resample weekly 
data.resample('W').sum().plot()
```
